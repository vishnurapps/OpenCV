Here we could see that the image is not displayed properly. This is because matplot expects colours in 
the order RGB and opencv is giving colors in the order BGR. We should fix this

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
new_img = cv2.resize(fix_img, (1000, 400))
```

```python
plt.imshow(new_img)
```

```python
width_ratio = .5
height_ratio = .5
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, width_ratio, height_ratio) #synthax to resize based in width and height ratio
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
new_img = cv2.flip(fix_img,0) #This is used to flip in horizontal axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,1) #This is used to flip in vertical axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,-1) #This is used to flip in vertical and horizontal axis
plt.imshow(new_img)
```
